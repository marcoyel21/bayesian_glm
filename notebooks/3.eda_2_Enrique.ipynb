# Transformación 1

```python
dicc_alc_cve = {
    'ALVARO OBREGON': '09010',
    'AZCAPOTZALCO': '09002',
    'BENITO JUAREZ': '09014',
    'COYOACAN': '09003',
    'CUAJIMALPA DE MORELOS': '09004',
    'CUAUHTEMOC': '09015',
    'GUSTAVO A. MADERO': '09005',
    'IZTACALCO': '09006',
    'IZTAPALAPA': '09007',
    'LA MAGDALENA CONTRERAS': '09008',
    'MIGUEL HIDALGO': '09016',
    'MILPA ALTA': '09009',
    'TLAHUAC': '09011', 
    'TLALPAN': '09012',
    'VENUSTIANO CARRANZA': '09017',
    'XOCHIMILCO': '09013'
}
```

```python
df['pct_enganche'] = 1-df['dat_monto_orig_cred']/df['dat_valor_vivienda_originacion']
df = df[
    (df.tasa_ponderada < 20)
    & (df['pct_enganche'] > 0)
]

df['fecha_originacion'] = pd.to_datetime(df.cve_periodo, format="%Y%m")

df = df.assign(clave=lambda x: x["dl_municipio"].map(dicc_alc_cve))
```

```python
df[
    df.dl_estado == "CIUDAD DE MÉXICO"
][["dl_municipio","clave"]].value_counts(dropna=False)
```

# Exploración 1

## Continuas

```python
cont_vars = [
    "dat_ingreso_mensual_bruto",
    "dat_valor_vivienda_originacion",
    "dat_ai_edad_acred",
    "dat_monto_orig_cred",
    "tasa_ponderada",
    "pct_enganche",
]

for var in cont_vars:
    df[var].hist(bins=35);
    plt.title(var);
    plt.show();
```

### Excluyendo outliers

```python
for var in cont_vars:
    quant_var_max = df[var].quantile(.975)
    quant_var_min = df[var].quantile(.015)
    df[
        (df[var] <= quant_var_max)
        & (df[var] >= quant_var_min) 
    ][var].hist(bins=35);
    plt.title(var);
    plt.show();
```

## Discretas

```python
cat_vars = [
    "nombre_publicacion",
    "sector",
    "dl_municipio",
    "tipo_acreditado",
    "dl_tipo_comprobacion",
    "dl_genero",
    "dl_destino_credito",
    "dl_producto_hipotecario",
    "Tipo de Crédito",
    "dl_segmento_vivienda",
    
]

for var in cat_vars:
    df[var].value_counts().head(20).plot(kind='barh');
    plt.title(var);
    plt.show();
```

```python
df[
    df.dat_monto_orig_cred < 15000000
].plot.scatter(x='dat_monto_orig_cred', y="tasa_ponderada");
```

```python
df_pivot_1 = df.pivot_table(
    index = "clave",
    aggfunc={
        'dat_monto_orig_cred':[sum, 'count','mean', 'median'],
        'dat_ai_edad_acred':['mean'],
        'dat_ingreso_mensual_bruto':['mean', 'median'],
        'tasa_ponderada':['mean'],
        'pct_enganche':['mean'],
    }
)

df_pivot_1.columns = [
    "edad_promedio",
    "ingreso_promedio",
    "mediana_ingreso",
    
    "creditos",
    "promedio_monto",
    "mediana_monto",
    "monto_originado",
    
    "enganche_promedio",
    "tasa_promedio",
    

]

df_pivot_1.head()
```

```python
df_map_plot = pd.concat(
    [
        cdmx,
        df_pivot_1
    ],
    axis=1
)

df_map_plot.tail()
```

```python
vars_to_plot = [
    "edad_promedio",
    "ingreso_promedio",
    "mediana_ingreso",
    "creditos",
    "promedio_monto",
    "enganche_promedio",
    "tasa_promedio",
]

for var in vars_to_plot:
    df_map_plot.plot(var, legend=True, cmap="OrRd");
    plt.title(var)
    plt.show();
```

```python
df['fecha_originacion'] = pd.to_datetime(df.cve_periodo, format="%Y%m")
df_pivot_1 = df.pivot_table(
    index = df.fecha_originacion.dt.to_period('M'),
    aggfunc={
        'dat_monto_orig_cred':[sum, 'count','mean', 'median'],
        'dat_ai_edad_acred':['mean'],
        'dat_ingreso_mensual_bruto':['mean', 'median']
    }
)

df_pivot_1.columns = [
    "edad_promedio",
    "ingreso_promedio",
    "mediana_ingreso",
    
    "creditos",
    "promedio_monto",
    "mediana_monto",
    "monto_originado",

]

df_pivot_1.head()
```
